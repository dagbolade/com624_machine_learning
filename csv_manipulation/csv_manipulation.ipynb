```python
from pandas import read_csv
from pandas import set_option
from matplotlib import pyplot
from pandas.plotting import scatter_matrix
import pandas as pd
import numpy as np
#from sklearn.impute import SimpleImputer


mental = read_csv("venv/MentalHQ2020.csv")


mental_num_variable = mental[['birthpositionfather', 'birthpositionmother', 'averagebreakperday']]

#print(mental_num_variable)

#using the simpleImputer funvtion to replace the NaN values
imputer = SimpleImputer(strategy = 'most_frequent')
imputer.fit(mental_num_variable)
mental_num_variable = imputer.transform(mental_num_variable)


# pass the numpy data to dataframe
mental_num_variable = pd.DataFrame(data=mental_num_variable, columns=['birthpositionfather', 'birthpositionmother', 'averagebreakperday'])
print('\n testing replacements')
print(mental_num_variable)

#getting basic stats for the numeric datasets
print(mental_num_variable.describe())

# print the histogram
mental_num_variable.hist()
pyplot.show()

# print the density plot
mental_num_variable.plot(kind = 'density', subplots = True, layout = (3,3), sharex = False)
pyplot.show()

# print the remaining va;ues that are alpha numeric by excluding the already declared numeric data
alpha_numeric = mental.loc[:, ~mental.columns.isin(['birthpositionfather', 'birthpositionmother', 'averagebreakperday'])]

print('\n', alpha_numeric.shape)
print(alpha_numeric.head(2))
```
